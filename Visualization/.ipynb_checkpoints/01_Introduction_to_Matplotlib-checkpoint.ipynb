```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3], [4,5,6])
plt.show()
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.xlabel("Years")
plt.ylabel("Runs")
plt.title("Sachin Tendulkar's Runs per Year")
plt.plot(years, runs)
plt.show() #While using jupyter ntebook we dont need to use it
```

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot(years, kohli, "ro--", label="Virat Kohli")
plt.plot(years, sehwag, color = "g", linestyle=":", linewidth=1.9, label="Virender Sehwag")

plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()
```
