```python
import pandas as pd
```

```python
data = {
    "Name": ["Alice", "Bob", "Charlile"],
    "Math": [85,78, 92],
    "Science": [90, 82, 89],
    "English": [88, 85, 94]
}
df = pd.DataFrame(data)
df
```

```python
df = df.melt(id_vars=["Name"], value_vars=["Math", "Science", "English"], var_name="Subject", value_name="Score")
df
```

```python
df = df.pivot(index="Name", columns="Subject", values="Score")
df
```
